## Using sklearn to transformer our data`

```python
data = pd.read_csv('car-sales-extended-missing-data.csv')
data
```

```python
data.isna().sum()
```

```python
x = data.drop('Price', axis=1)
y =data.Price
```

```python
fy = pd.DataFrame(y, columns=['Price'])
fy
```

```python
x.isna().sum()
x
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

cat_inputer = SimpleImputer(strategy='constant', fill_value='missing')
odo_inputer = SimpleImputer(strategy='mean')
door_inputer = SimpleImputer(strategy='constant', fill_value=4)

# create the features
cat_features = ['Make', 'Colour']
odo_features = ['Odometer (KM)']
door_feat = ['Doors']

# fit em
imputer = ColumnTransformer([('cat inputer', cat_inputer, cat_features),
                             ('odo features', odo_inputer, odo_features),
                             ('door feat', door_inputer, door_feat)    
])

filled_x = imputer.fit_transform(x)
filled_x
```

```python
car_sales_data = pd.DataFrame(filled_x, columns=['Make','Colour', 'Odometer (KM)', 'Doors'])
```

```python
dummies = pd.get_dummies(car_sales_data)
```

```python
car_sales_data.isna().sum()
```

```python
y.isna().sum()
```

```python
price_inputer = SimpleImputer(strategy='mean')
fet= ['Price']
y_imputer = ColumnTransformer([('price', price_inputer, fet)])
filled_y = y_imputer.fit_transform(fy)
```

```python
filled_y
y = pd.DataFrame(filled_y, columns=['Price'])
```

```python
y_num = pd.DataFrame.to_numpy(y)
y_num
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x_train, x_test, y_train, y_test = train_test_split(dummies, y_num, train_size=0.8)
model = RandomForestRegressor()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

## GETTING REAL DATA SETS FROM THE INTERNET
1) choosing which estimators to use

```python
from sklearn.datasets import fetch_california_housing
house = fetch_california_housing()
```

```python
house
```

```python
housing = pd.DataFrame(house['data'], columns=house['feature_names'])
```

```python
housing
```

```python
housing['target'] = house['target']
```

```python
housing
```

```python
np.random.seed(42)
x = housing.drop('target', axis=1)
y = housing['target']
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y, train_size=0.8)
```

using Ridge

```python
from sklearn.linear_model import Ridge
model = Ridge()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

using lasso

```python
from sklearn.linear_model import Lasso
model = Lasso()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

using support vector machines

```python
from sklearn import svm
model = svm.SVR()
model.fit(x_train,y_train)
model.score(x_test, y_test)
```

using ensemnble method

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
y_preds = model.predict(x_test)
```
