```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Scraped_dataset.csv', parse_dates=['Date of Booking', 'Date of Journey'], infer_datetime_format=True )
```

```python
df
```

```python
df.sort_values(['Price'], ascending=False)
```

```python
df.info()
```

```python
df.Price = df.Price.str.replace(',', '', regex=True).astype(int)
```

```python
df.info()
```

```python
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
df[['Airline', 'Serial Number', 'Class']] = df['Airline-Class'].str.split('\n', expand=True)
df.drop('Airline-Class', axis=1, inplace=True)
```

```python
df.head()
```

```python
df[['Time Of Departure', 'Departure Airport']] = df['Departure Time'].str.split('\n', expand=True)
df.drop('Departure Time', axis=1, inplace=True)
```

```python
df['Date_of_journey'] = df['Date of Journey']
df['Date_of_bookin'] = df['Date of Booking']
```

```python
df.drop('Date of Journey', axis=1, inplace=True)
df.drop('Date of Booking', axis=1, inplace=True)
```

```python
df.head()
```

```python
df[['Time Of Arrival', 'Arrival Airport']] = df['Arrival Time'].str.split('\n', expand=True)
df.drop('Arrival Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Day_interval'] = (df['Date_of_journey'] - df['Date_of_bookin']).astype(str)
```

```python
df.head()
```

```python
df.Day_interval = df.Day_interval.str.replace(' days', '', regex=True).astype(int)
```

```python
df.head()
```

```python
df['Year'] = df.Date_of_journey.dt.year
df['Month'] = df.Date_of_journey.dt.month
df['DayofMonth'] = df.Date_of_journey.dt.day
df['Week'] = df.Date_of_journey.dt.isocalendar().week
df['Dayofweek'] = df.Date_of_journey.dt.day_of_week
```

```python
data = df.copy()
```

```python
data.drop('Date_of_bookin', axis=1, inplace=True)
```

```python
data
```

```python
def duration_conversion(time_str):
    hours, minutes = time_str.split('h ')
    return int(hours) + int(minutes[:-1])/60

data['flight_duration'] = data.Duration.apply(duration_conversion)
```

```python
data
```

```python
data.drop('Duration', axis=1, inplace=True)
```

```python
data['DayOfJourney'] = data.Date_of_journey.dt.day_name()
```

```python
data.drop('Date_of_journey', axis=1, inplace=True)
```

```python
data
```

```python
data.info()
```

```python
def to_mins(time_str):
    hours, minutes = map(int, time_str.split(':'))
    return hours * 60 + minutes


data['Arrival_time_in_mins'] = data['Time Of Arrival'].apply(to_mins)
data['Departure_time_mins'] = data['Time Of Departure'].apply(to_mins)
```

```python
data
```

```python
data.drop('Time Of Departure', axis=1, inplace=True)
data.drop('Time Of Arrival', axis=1, inplace=True)
```

```python
data.info()
```

```python
data
```

```python
for column, content in data.items():
    if pd.api.types.is_string_dtype(content):
        data[column] = content.astype('category').cat.as_ordered()
    if pd.api.types.is_categorical_dtype(content):
        data[column] = pd.Categorical(content).codes+1
```

```python
data.info()
```

# Machine Learning

Our Data is Continous so we will use the different regression estimator from the sklearn map
* https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html

```python
from sklearn.model_selection import train_test_split
x = data.drop('Price', axis=1)
y = data.Price

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

### Using SGD Regressor

```python
%time
from sklearn.linear_model import SGDRegressor
reg = SGDRegressor()
reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```

### Using Linear Regressor

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```

### Using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```

Random Forest is the way then
