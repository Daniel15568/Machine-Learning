```python
import pandas as pd
import numpy as np
```

```python
rawdata = pd.read_csv('cars.csv')
pd.options.display.max_columns=200
data = rawdata.drop('location_region', axis=1)
```

```python
data.isna().sum()
```

```python
data.dropna(subset='engine_capacity', inplace=True)
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
data.head()
```

```python
data.year_produced = data['year_produced'].astype(str)
year = pd.get_dummies(data.year_produced, prefix='Year')
data = pd.concat([data, year], axis=1)
data.drop('year_produced', axis=1)
```

```python
x= data.drop('price_usd', axis=1)
y= data.price_usd
```

```python
x = data.drop('model_name', axis=1)
x
```

```python
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

columns = ('manufacturer_name', 'transmission','body_type','state', 'engine_type','drivetrain',
           'color','engine_fuel','feature_0', 'feature_1','feature_2', 
           'feature_3','feature_4','feature_5','feature_6','feature_7','feature_8', 'feature_9', 'has_warranty', 
           'engine_has_gas', 'is_exchangeable')

cat_transform = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])

processor = ColumnTransformer(transformers=[('cat', cat_transform, columns)])

model = Pipeline(steps=[('processor', processor), 
                       ('model', RandomForestRegressor(n_estimators=100))])
```

```python
# new_ex = pd.get_dummies(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
