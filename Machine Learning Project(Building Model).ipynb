# Import all Neccessary Models and Libraries for Modelling

```python
np.random.seed(42)
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

# create a function to run the models

```python
models = {'Logistic Regression':LogisticRegression(),
          'Random Forest': RandomForestClassifier(),
          'KNN': KNeighborsClassifier()}

def run_models(model, x_train,x_test,y_train,y_test):    
    
    np.random.seed(10)
    
    model_score = {}
    
    for name, model in models.items():
        model.fit(x_train,y_train)
        model_score[name] = model.score(x_test, y_test)
    return model_score
```

# MODELLING

```python
data
```

```python
x = data.drop('outcome', axis=1)
y = data.outcome
```

```python
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
job = run_models(models, x_train, x_test, y_train, y_test)
```

```python
job
```

```python
Original_result = pd.DataFrame(job, index=['Accuracy'])
```

```python
Original_result.T.plot(kind='bar');
```

# TUNING HYPERPARAMETERS

# Kneighbors

```python
test_scores = []

knn = KNeighborsClassifier()
neigbors = range(1,21)

for i in neigbors:
    knn.set_params(n_neighbors=i)
    knn.fit(x_train, y_train)
    test_scores.append(knn.score(x_test, y_test))
```

```python
test_scores
```

```python
plt.plot(test_scores);
```

```python
KNN = max(test_scores)
```

# Using Randomized Search CV

```python
log_reg_par = {'C': np.logspace(-4, 4, 20),
               'solver': ['liblinear', 'sag', 'saga']}

ran_for_par = {'n_estimators': np.arange(10, 1000, 100),
              'max_depth': [None, 2, 10] ,
               'max_features': ['sqrt', 'log2'],
               'min_samples_split': np.arange(2, 10, 2),
               'min_samples_leaf' : np.arange(1, 10, 2)}

rf_clf = RandomForestClassifier()
lr_clf = LogisticRegression()

log_clf= RandomizedSearchCV(estimator=lr_clf, param_distributions=log_reg_par, n_iter=50, verbose=2,
                    cv=5)
ran_clf = RandomizedSearchCV(rf_clf, param_distributions=ran_for_par, n_iter=20, cv=5, verbose=2)
```

 # CLASSIFICATION REPORT using cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_precision = cross_val_score(clf, x, y, cv=5, 
                            scoring='precision')
cv_precision = np.mean(cv_precision)
```

```python
cv_recall = cross_val_score(clf, x, y, cv=5, 
                            scoring='recall')
cv_recall = np.mean(cv_recall)
```

```python
cv_f1_score = cross_val_score(clf, x, y, cv=5, 
                            scoring='f1')
cv_f1_score = np.mean(cv_f1_score)
```

```python
cv_accuracy = cross_val_score(clf, x, y, cv=5, 
                            scoring='accuracy')
cv_accuracy = np.mean(cv_accuracy)
```

```python
data = {'accuracy':cv_accuracy,
        'F1 Score':cv_f1_score,
        'recall': cv_recall,
        'precision': cv_precision}

df = pd.DataFrame(data, index=['scores'])
```
